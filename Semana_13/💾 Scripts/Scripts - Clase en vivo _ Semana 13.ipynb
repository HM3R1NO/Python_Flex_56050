```python
from django.contrib.auth.models import User
class Avatar(models.Model):
    user = models.OneToOneField(User, on_delete=models.CASCADE)
    imagen = models.ImageField(upload_to='avatares', null=True, blank = True)
    descripcion = models.TextField(null=True)
    link = models.URLField(null=True)

    def __str__(self):
        return f"{self.user} - {self.imagen}"
```

```python
def multiplo(numero):     # Primero declaramos una funcion condicional
    if numero % 5 == 0:   # Comprobamos si un numero es multiplo de 5
        return True       # Solo devolvemos True si lo es
   
numeros = [2, 5, 10, 23, 50, 33]


numeros_filtrados = filter(multiplo, numeros)
print(numeros_filtrados)            # Como veran esto devuelve un objeto filter
print(list(numeros_filtrados))      # Asi que lo pasaremos a lista
```

```python
def doblar(numero):
    return numero*2


numeros = [2, 5, 10, 23, 50, 33]
numeros_mapeados = map(doblar, numeros)
print(numeros_mapeados)            # Como veran esto devuelve un objeto map
print(list(numeros_mapeados))      # Asi que lo pasaremos a lista
```
